```python
import pandas as pd

path = r"C:\Users\Vishal(CloudPlatform\Downloads\logs\logs\2026-04-24\consolidated-backend.log"

logs = []

with open(path, "r", encoding="utf-8") as f:

    for line in f:

        parts = line.strip().split(" | ")

        # Ensure proper parsing
        if len(parts) >= 6:

            logs.append({
                "timestamp": parts[0],
                "log_level": parts[1],
                "process_id": parts[2],
                "logger_name": parts[3],
                "source": parts[4],
                "message": " | ".join(parts[5:])
            })

# Create DataFrame
df = pd.DataFrame(logs)

# =========================
# CLEAN DATA
# =========================

# Remove extra spaces
for col in df.columns:
    df[col] = df[col].astype(str).str.strip()

# Convert timestamp
df["timestamp"] = pd.to_datetime(
    df["timestamp"],
    format="%Y-%m-%d %H:%M:%S,%f",
    errors="coerce"
)

# Convert process_id
df["process_id"] = pd.to_numeric(
    df["process_id"],
    errors="coerce"
)

# Remove invalid rows
df = df.dropna(subset=["timestamp"])

# =========================
# TESTING
# =========================

print("\nDATAFRAME:")
print(df.head())

print("\nLOG LEVEL COUNTS:")
print(df["log_level"].value_counts())

print("\nONLY ERRORS:")
errors = df[df["log_level"] == "ERROR"]
print(errors)

print("\nMODULAR AGENT LOGS:")
print(df[df["logger_name"] == "modular_agents.system"])

print("\nUNIQUE LOG LEVELS:")
print(df["log_level"].unique())
```
